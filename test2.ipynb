```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
V = pd.read_csv('lab4.csv', delimiter = ';')
# V = V.applymap(lambda x: float(str(x).replace(',', '.')) if isinstance(x, str) else x)
# V.to_csv('lab4.csv', index = False)
V = V.iloc[::-1].reset_index(drop=True)

ys, xs = V.shape

x = np.arange(xs)
y = np.arange(ys)
X, Y = np.meshgrid(x,y)

plt.figure(figsize = (20,10))
plt.contourf(X,Y, V, levels = 50, cmap = 'Greens')
plt.title('Napięcie wewnątrz kondensatora płaskiego')
plt.xlabel('x [cm]')
plt.ylabel('y [cm]')
plt.colorbar(label = 'Voltage [V]')
```

```python
newV = V.iloc[1:-1]
y2, x2 = newV.shape

newX, newY = np.meshgrid(np.arange(x2), np.arange(y2))

plt.figure(figsize = (20,10))
plt.contour(X,Y, V, levels = 9, cmap = 'Greens')
plt.title('Napięcie wewnątrz kondensatora płaskiego')
plt.xlabel('x [cm]')
plt.ylabel('y [cm]')
plt.colorbar(label = 'Voltage [V]')
```

```python
Ey, Ex = np.gradient(-newV)
magnitude = np.sqrt(Ex ** 2 + Ey ** 2)
plt.figure(figsize = (20,10))
plt.title('')
plt.xlabel('x [cm]')
plt.xlabel('y [cm]')
plt.quiver(newX, newY, Ex, Ey, magnitude, color = 'green', scale = 20)
plt.ylim(-1, 4.5)
```
